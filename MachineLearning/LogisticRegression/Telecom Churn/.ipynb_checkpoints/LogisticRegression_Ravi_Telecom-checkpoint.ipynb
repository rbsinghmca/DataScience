# Plotting ROC Curve

Trade off between sensitivity and specifity(If sensitivity increase then specifity decrease)<br>
The closer the curve follow left-hand border then top border of ROC space then more accurate the test <br>
The closer the curve comes to 45-degree ROC space, the less accurate the test

```python
from sklearn import metrics
```

```python
def draw_roc(actuals, probs):
    fpr, tpr, thresholds = metrics.roc_curve(actuals, probs, drop_intermediate=False)
    auc_score = metrics.roc_auc_score(actuals, probs)
    plt.figure(figsize = (5,5))
    plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score)
    plt.plot([0,1], 'k--')
    plt.xlim([0.0,1.0])
    plt.ylim([0.0,1.05])
    plt.xlabel('False Positive Rate or 1-specifity or 1-TNR')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristics example')
    plt.legend(loc='lower right')
    plt.show()
    
    return None
draw_roc(y_train_pred_final['Churn'], y_train_pred_final['Churn_Prob'])
```

# Finding optimal Curve

```python
# Create columns with different probability curve
numbers = [float(x)/10 for x in range(10)]
for i in numbers:
    y_train_pred_final[i] = y_train_pred_final['Churn_Prob'].map(lambda x: 1 if x > i else 0)
y_train_pred_final.head()
```

##### Calculate accuracy, sentivity and specifity for various cuttoffs

```python
cuttoffs_df = pd.DataFrame(columns=['prob_cut-off', 'accuracy', 'sensitivity', 'specifity'])
from sklearn.metrics import confusion_matrix
# TN = confusion[0,0]
# FP = confusion[0,1]
# FN = confusion[1,0]
# TP = confusion[1,1]
numbers = [float(x)/10 for x in range(10)]
for i in numbers:
    cm = confusion_matrix(y_train_pred_final['Churn'],y_train_pred_final[i]) # y_train_pred_final['Churn']
    total=sum(sum(cm))
    acc = (cm[0,0]+cm[1,1])/total
    sen = cm[1,1]/(cm[1,1]+cm[1,0])
    spec = cm[0,0]/(cm[0,0]+cm[0,1])
    cuttoffs_df.loc[i] = [i, acc, sen, spec]
cuttoffs_df
```

#### Calculate Accuracy, sensitivity, specifity for various cutt-offs

```python
cuttoffs_df.plot.line(x='prob_cut-off', y=['accuracy','sensitivity','specifity'])
plt.show()
```

In the above graph, the optimal cuttoffs is 0.3 as a cutoff probability
