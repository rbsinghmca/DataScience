# Data Cleaning

```python
# Checking null values percentage in each columns
100*(telecom.isnull().sum(axis=0)/len(telecom))
```

```python
# Dropping columns which has high percentage of missing data
cols = telecom.columns
dropped_columns = []
for column in cols:
    if telecom[column].isnull().sum()/len(telecom) > .7:
        dropped_columns.append(column)
        telecom = telecom.drop(column, axis=1)
print('Following columns were dropped ', dropped_columns)
```

```python
# Checking null values percentage in each columns again
100*(telecom.isnull().sum(axis=0)/len(telecom))
```

# Data Preparation

```python
# Dropping some variable based high correlation
# Create correlation matrix
corr_matrix = telecom_hvc[vars_num].corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.85)]
print('Dropped Columns are ', to_drop)
# Dropping highly correlated columns(> 85)
telecom_hvc.drop(to_drop, axis=1, inplace = True)
```

```python
# Reducing the number of category in the sachet field to minimise the number of feature after dummification
telecom_hvc['sachet_2g_6'].replace([18.0,19.0,16.0,22.0,25.0,24.0,20.0,21.0,32.0,34.0,39.0,27.0,30.0],31, inplace=True)
telecom_hvc['sachet_2g_7'].replace([18.0,19.0,16.0,22.0,25.0,24.0,20.0,21.0,32.0,34.0,39.0,27.0,30.0],31, inplace=True)
telecom_hvc['sachet_2g_8'].replace([18.0,19.0,16.0,22.0,25.0,24.0,20.0,21.0,32.0,34.0,39.0,27.0,30.0],31, inplace=True)
telecom_hvc['sachet_3g_6'].replace([10.0,11.0,12.0,13.0,14.0,15.0,16.0,17.0,18.0,19.0,16.0,22.0,25.0,24.0,20.0,21.0,32.0,34.0,39.0,27.0,30.0,31.0,32.0,33.0,34.0,35.0],36, inplace=True)
telecom_hvc['sachet_3g_7'].replace([10.0,11.0,12.0,13.0,14.0,15.0,16.0,17.0,18.0,19.0,16.0,22.0,25.0,24.0,20.0,21.0,32.0,34.0,39.0,27.0,30.0,31.0,32.0,33.0,34.0,35.0],36, inplace=True)
telecom_hvc['sachet_3g_8'].replace([10.0,11.0,12.0,13.0,14.0,15.0,16.0,17.0,18.0,19.0,16.0,22.0,25.0,24.0,20.0,21.0,32.0,34.0,39.0,27.0,30.0,31.0,32.0,33.0,34.0,35.0],36, inplace=True)
```

### Hyper-parameter tuning for the Decision Tree

```python
from sklearn.model_selection import GridSearchCV
```

```python
dt = DecisionTreeClassifier(class_weight='balanced',random_state=42)
```

```python
params = {
    "max_depth": [2,3,5,10,20],
    "min_samples_leaf": [5,10,20,50,100,500]
}
```

```python
grid_search = GridSearchCV(estimator=dt,
                           param_grid=params,
                           cv=4,
                           n_jobs=-1, verbose=1, scoring="accuracy")
```

```python
%%time
grid_search.fit(df_train_pca, y_train)
```

```python
grid_search.best_score_
```

```python
dt_best = grid_search.best_estimator_
dt_best
```

```python
plot_roc_curve(dt_best, df_train_pca, y_train)
plt.show()
```
