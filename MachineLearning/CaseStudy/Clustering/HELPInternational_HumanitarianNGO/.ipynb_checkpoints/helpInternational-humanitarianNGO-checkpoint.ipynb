# Step 4 Model Building

## Hopkins Statistics 
- To verify that is the dataset clusterable or not

```python
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan
 
def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n) 
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
 
    rand_X = sample(range(0, n, 1), m)
 
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0
 
    return H
```

```python
hs = []
for i in range(10):
    hs.append(hopkins(ngo_scaled))
    
print(list(map(lambda x: round(x,2),hs)))
print('Average of the 10 different run of hopkin score is {0}'.format(np.mean(hs)))
```

The data has well custer tendency as average of ten hopkin scores is more than 85 percentage

# Builing Model Using KMeans

```python
# Building model by initializing of k with random value 5
kmeans = KMeans(n_clusters=5, random_state=10)
kmeans.fit(ngo_scaled)
kmeans.labels_
```

```python
# Checking of the number of countiries in each cluster
np.unique(kmeans.labels_,return_counts=True)
```

## Optimal Number of Clusters

### Checking Silhouette Score
- Checking the optimal number of clusterings in the data set

```python
# Silhouette score
ss = []
for i in range(2,12):
    kmean = KMeans(n_clusters=i).fit(ngo_scaled)
    ss.append([i,silhouette_score(ngo_scaled,kmean.labels_)])
temp = pd.DataFrame(ss)
plt.plot(temp[0],temp[1])
plt.show()
```

- Here maximum height/location after number number 2 is number 3. Number of clusters 2 would not be better for clusterting because these countries are spread across different geographical location and in different natures so taking next highest that is 3 
- Let's see also elbow curve to decide the optimal number of clusters

# Elbow Curve
Also checking the elbow curve for optimal numbers of clustering for cross validation with silhouette score

```python
# Elbow Curve
ssd = []
for i in range(2,12):
    kmean = KMeans(n_clusters=i).fit(ngo_scaled)
    ssd.append([i,kmean.inertia_])
temp = pd.DataFrame(ssd)
plt.plot(temp[0],temp[1])
plt.show()
```

- Here is also a bend at 3. So considering number of clusters as per both silhouette score and elbow curve.
- Finally optimal number of clusters taken k =3

### Builing Final Model using KMeans by taking n_clusers =3

```python
# Building model using KMeans 
k = 3
kmeans = KMeans(n_clusters = k, random_state = 25)
kmeans.fit(ngo_scaled)
label = kmeans.labels_
```

# Hierarchical Clustering 

```python
# Checking data before clustering
ngo_scaled.head()
```

```python
# Single Linkage
plt.figure(figsize=(15,8))
mergings = linkage(ngo_scaled, method='single', metric='euclidean')
dendrogram(mergings)
plt.show()
```

It is difficult to ust and visualise in single linkage. Checking on complete linkage

```python
# Complete Linkage
plt.figure(figsize=(15,8))
mergings = linkage(ngo_scaled, method='complete', metric='euclidean')
dendrogram(mergings)
plt.show()
```

Method Complete is better to visualise. By visualsing,either we can cut off 3 clusters or 5 clusters.
But I would like to go for 3 as per silhouette score and elbow curve

```python
# Cutting the dendrogram at just below the 10 means number of clusters = 3
label_by_hier = cut_tree(mergings, n_clusters =3).reshape(-1,)
```

```python
# Adding anothe label to original dataframe for hierachical result
ngo['label_by_hier'] = label_by_hier
```
