# Step 2 Data Preparation

```python
# Having a look at data
bike.head()
```

### Dummifying Categorical Variables

```python
# yr, holiday, workingday has binary fields which are good for regression.
# mnth, season, weekday, weathersit are categorical columns and these variables has no any order kind of relation with target column so dummifying these.
```

```python
bike.info()
```

```python
# Mapping the mnth column with Month names
lst_mnth={
1:'January',
2:'February',
3:'March',
4:'April',
5:'May',
6:'June',
7:'July',
8:'August',
9:'September',
10:'October',
11:'November',
12:'December'}
bike['mnth'] = bike['mnth'].map(lst_mnth)
```

```python
# Mapping Season column
lst_season={1:'spring', 2:'summer', 3:'fall', 4:'winter'}
bike['season'] = bike['season'].map(lst_season)
```

```python
# Mapping weekday column
lst_weekday={
0:'Sunday',
1:'Monday',
2:'Tuesday',
3:'Wednesday',
4:'Thursday',
5:'Friday',
6:'Saturday'
}
bike['weekday'] = bike['weekday'].map(lst_weekday)
```

```python
# Mapping weathersit column
lst_weathersit={
1:'Clear',
2:'Mist_Clody',
3:'Light_Snow_Rain',
4:'Heavy_Rain_IcePallets'
}
bike['weathersit'] = bike['weathersit'].map(lst_weathersit)
```

```python
bike.head()
```

```python
# Dummifying Variables season, mnth, weekday, weathersit then dropping these(original) variables
vars_dummies = ['season', 'mnth', 'weekday', 'weathersit']
for i in vars_dummies:
    j = i  # drop this variable from the dataset once it is dummified
    i = pd.get_dummies(bike[i], drop_first = True)
    bike = pd.concat([bike,i], axis = 1)
    bike.drop(j, axis = 1, inplace = True)
```

```python
# Checking all the variables created
pd.options.display.max_columns = 35
bike.head()
```

```python
# Checking its data types
bike.info()
```

# Step 3 Model Building

```python
# Importing RFE and LinearRegression from sklearn
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Putting data for X and Y axis for regression
y_train = df_train.pop('cnt')
x_train = df_train
```

```python
# Linear Regression
lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
# Running RFE for 15 features
rfe = RFE(lm, 10)
rfe = rfe.fit(x_train,y_train)
```

```python
# Ranking features with support made by using RFE 
list(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

```python
# Features selected by RFE
cols = x_train.columns[rfe.support_]
cols
```

```python
# Features not selected by RFE
x_train.columns[~rfe.support_]
```

```python
# Taking 15 columns supported by RFE for regression
x_train_rfe = x_train[['yr', 'holiday', 'workingday', 'temp', 'hum', 'windspeed', 'spring',
       'summer', 'winter', 'January', 'July', 'September', 'Saturday',
       'Light_Snow_Rain', 'Mist_Clody']]
```

```python
x_train_rfe.shape
```

```python
# Importing statsmodels for regression
import statsmodels.api as sm
```

### Model 1

```python
# Adding constant to X variable as statsmobels this do not support intercept by default
x_train_rfe_sm = sm.add_constant(x_train_rfe)
```

```python
# Fit regressional line
lr1 = sm.OLS(y_train,x_train_rfe_sm).fit()
lr1.summary()
```

```python
# Checking correlation of features selected by RFE with target column. 
# Also to check impact of different features on target.
plt.figure(figsize = (15,10))
sns.heatmap(bike[['cnt','yr', 'holiday', 'workingday', 'temp', 'hum', 'windspeed', 'spring',
       'summer', 'winter', 'January', 'July', 'September', 'Saturday',
       'Light_Snow_Rain', 'Mist_Clody']].corr(), cmap='GnBu', annot=True)
plt.show()
```

### Model 2

```python
# Dropping the variable January as it has negative coefficient and is insignificant as it has high p-value
x_train_rfe = x_train[['yr', 'holiday', 'workingday', 'temp', 'hum', 'windspeed', 'spring',
       'summer', 'winter','July', 'September', 'Saturday',
       'Light_Snow_Rain', 'Mist_Clody']]
x_train_rfe_sm = sm.add_constant(x_train_rfe)
lr2 = sm.OLS(y_train,x_train_rfe_sm).fit()
lr2.summary()
```

### All the columns have p-value > .05 so checking VIFs

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['Feature'] = x_train_rfe.columns
vif['Response'] = [variance_inflation_factor(x_train_rfe.values, i) for i in range(x_train_rfe.shape[1])]
vif['Response'] = round(vif['Response'],1)
vif = vif.sort_values(by = 'Response', ascending = False)
vif
```

### Model 3

```python
# Dropping hum column as it has very high vif > 10 and rebuilding the model
x_train_rfe = x_train[['yr', 'holiday', 'workingday', 'temp','windspeed', 'spring',
       'summer', 'winter', 'July', 'September', 'Saturday',
       'Light_Snow_Rain', 'Mist_Clody']]
x_train_rfe_sm = sm.add_constant(x_train_rfe)
lr3 = sm.OLS(y_train,x_train_rfe_sm).fit()
lr3.summary()
```

All the features have p-value < .05

```python
# Checking VIF again
vif = pd.DataFrame()
vif['Feature'] = x_train_rfe.columns
vif['Response'] = [variance_inflation_factor(x_train_rfe.values, i) for i in range(x_train_rfe.shape[1])]
vif['Response'] = round(vif['Response'],1)
vif = vif.sort_values(by = 'Response', ascending = False)
vif
```

### Model 4

```python
## Dropping temp column as it high VIF > 5
x_train_rfe = x_train[['yr', 'holiday', 'workingday','windspeed', 'spring',
       'summer', 'winter', 'July', 'September', 'Saturday',
       'Light_Snow_Rain', 'Mist_Clody']]
x_train_rfe_sm = sm.add_constant(x_train_rfe)
lr4 = sm.OLS(y_train,x_train_rfe_sm).fit()
lr4.summary()
```

### Model 5

```python
## Dropping July as it is insignificant

x_train_rfe = x_train[['yr', 'holiday', 'workingday','windspeed', 'spring',
       'summer', 'winter','September', 'Saturday',
       'Light_Snow_Rain', 'Mist_Clody']]
x_train_rfe_sm = sm.add_constant(x_train_rfe)
lr5 = sm.OLS(y_train,x_train_rfe_sm).fit()
lr5.summary()
```

### Model 6

```python
## Dropping holiday as it is insignificant
x_train_rfe = x_train[['yr','workingday','windspeed', 'spring',
       'summer', 'winter','September', 'Saturday',
       'Light_Snow_Rain', 'Mist_Clody']]
x_train_rfe_sm = sm.add_constant(x_train_rfe)
lr6 = sm.OLS(y_train,x_train_rfe_sm).fit()
lr6.summary()
```

# Step 5 Residual Analysis

### Distribution of error terms

```python
# Cheching the distribution of error term
plt.figure()
y_train_pred = lr11.predict(x_train_rfe_sm)
sns.distplot(y_train - y_train_pred)
plt.suptitle('Error Terms', fontsize=20)  # plotting heading
plt.xlabel('Error', fontsize=20)          # Xlabel
```

### Errors are normally distribured here with mean 0. So everything seems to be fine

### Checking residuals pattern

```python
# Actual vs Predicted
c = [i for i in range(0,len(x_train),1)]
plt.plot(c,y_train, color="blue")
plt.plot(c,y_train_pred, color="red")
plt.suptitle('Actual vs Predicted', fontsize = 15)
plt.xlabel('Index')
plt.ylabel('Demands')
plt.show()
```

Actual and Predicted result following almost the same pattern so this model seems ok

```python
# Error Terms
c = [i for i in range(0,len(x_train),1)]
plt.plot(c,y_train-y_train_pred)
plt.suptitle('Error Terms', fontsize = 15)
plt.xlabel('Index')
plt.ylabel('y_train-y_train_pred')
plt.show()
```

Here,If we see the error terms are independent of each other.

### Linearity Check

```python
# scatter plot for the check
residual = (y_train - y_train_pred)
plt.scatter(y_train,residual)
plt.ylabel("y_train")
plt.xlabel("Residual")
plt.show()
```
